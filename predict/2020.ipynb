```python
def build_row(df, team_one, team_two):
    res = []
    
    team_one_row = dict(df[df['Team'] == team_one].iloc[0,:])
    team_two_row = dict(df[df['Team'] == team_two].iloc[0,:])
    
    curr_row = dict()
    for k, v in team_one_row.items():
        curr_row[k + '_team_one'] = v
    for k, v in team_two_row.items():
        curr_row[k + '_team_two'] = v
    
    res = pd.DataFrame(data=[curr_row])
    cols_to_drop = ['Team_team_one', 'Team_team_two', 'year_team_one', 'year_team_two']
    return (res, res.drop(columns=cols_to_drop).to_numpy())
```

```python
def gen_matchups(data):
    res = []
    first = 0
    last = len(data) - 1
    while first<last:
        res.append((data[first], data[last]))
        first += 1
        last -= 1
    return res
```

```python
def predict(model, df, team_one, team_two):
    _, row = build_row(df, team_one, team_two)
    
    res = model.predict(row)
    prob = model.predict_proba(row)
    winner = team_one if res == 1 else team_two
    
    return (winner, prob)
```

## Load the Models

```python
names = ['DataInvertMinMax', 'DataInvertStandard', 'DataMinMax', 'DataStandard']
```

```python
def load_models(prefix, names):
    res = []
    for name in names:
        model = joblib_load(f'{prefix}{name}.pkl')
        res.append(model)
    return res
```

```python
forests = load_models('models/forest/', names)
```

```python
logregs = load_models('models/logreg/', names)
```

```python
mlps = load_models('models/mlp/', names)
```

```python
def predict_all(df, team_one, team_two):
    res = []
    total_prob_one = 0
    total_prob_two = 0
    
    for forest in forests:
        curr = predict(forest, df, team_one, team_two)
        total_prob_one += curr[1][0][1]
        total_prob_two += curr[1][0][0]
        res.append(curr)
        
        curr = predict(forest, df, team_two, team_one)
        total_prob_one += curr[1][0][0]
        total_prob_two += curr[1][0][1]
        res.append(curr)
    for logreg in logregs:
        curr = predict(logreg, df, team_one, team_two)
        total_prob_one += curr[1][0][1]
        total_prob_two += curr[1][0][0]
        res.append(curr)
        
        curr = predict(logreg, df, team_two, team_one)
        total_prob_one += curr[1][0][0]
        total_prob_two += curr[1][0][1]
        res.append(curr)
    for mlp in mlps:
        curr = predict(mlp, df, team_one, team_two)
        total_prob_one += curr[1][0][1]
        total_prob_two += curr[1][0][0]
        res.append(curr)
        
        curr = predict(mlp, df, team_two, team_one)
        total_prob_one += curr[1][0][0]
        total_prob_two += curr[1][0][1]
        res.append(curr)
    
    teams_pred = [item[0] for item in res]

    count_one = teams_pred.count(team_one)
    count_two = teams_pred.count(team_two)

    prob_one = total_prob_one / len(res)
    prob_two = total_prob_two / len(res)
    
    print (f'{count_one}, {count_two}, {prob_one}, {prob_two}')
    
    prob_pair = [prob_one, prob_two]
    if prob_one > prob_two:
        return (team_one, prob_pair)
    elif prob_two > prob_one:
        return (team_two, prob_pair)
    else:
        return ('WTF', prob_pair)
#     if count_one > count_two:
#         if prob_one > prob_two:
#             return (team_one, prob_pair)
#         else:
#             if abs(count_one - count_two) <= 2:
#                 return (team_two, prob_pair)
#             else:
#                 return (team_one, prob_pair)
#     elif count_two > count_one:
#         if prob_two > prob_one:
#             return (team_two, prob_pair)
#         else:
#             if abs(count_one - count_two) <= 2:
#                 return (team_one, prob_pair)
#             else:
#                 return (team_two, prob_pair)
#     else:
#         if prob_one > prob_two:
#             return (team_one, [prob_one, prob_two])
#         elif prob_two > prob_one:
#             return (team_two, [prob_one, prob_two])
#         else:
#             return ('WTF', [prob_one, prob_two])
```

```python
class Tournament:
    def __init__(self, sixteen_teams):
        self.df = total_data[total_data['Team'].isin(sixteen_teams)]
        
        for idx, row in self.df.iterrows():
            curr_team = row['Team']
            team_idx = sixteen_teams.index(curr_team)
            self.df.at[idx, 'Seed'] = seeds[team_idx]
        
        self.first_round = gen_matchups(sixteen_teams)
        
    def run(self):
        print ('First Round:')
        self.first_round_results = []
        for matchup in self.first_round:
            res = predict_all(self.df, matchup[0], matchup[1])
            print (f'{matchup[0]} - {matchup[1]} -> {res}')
            
            self.first_round_results.append(res[0])
        print ('')
        
        print ('Second Round:')
        second_round = [(0, 7), (1, 6), (2, 5), (3, 4)]
        self.second_round = [(self.first_round_results[item[0]], self.first_round_results[item[1]]) for item in second_round]
        self.second_round_results = []
        for matchup in self.second_round:
            res = predict_all(self.df, matchup[0], matchup[1])
            print (f'{matchup[0]} - {matchup[1]} -> {res}')
            
            self.second_round_results.append(res[0])
        print ('')
        
        print ('Semis:')
        semis = [(0, 3), (1, 2)]
        self.semis = [(self.second_round_results[item[0]], self.second_round_results[item[1]]) for item in semis]
        self.semi_results = []
        for matchup in self.semis:
            res = predict_all(self.df, matchup[0], matchup[1])
            print (f'{matchup[0]} - {matchup[1]} -> {res}')
            
            self.semi_results.append(res[0])
        print ('')
        
        print ('Championship:')
        self.championship = (self.semi_results[0], self.semi_results[1])
        res = predict_all(self.df, self.championship[0], self.championship[1])
        print (f'{self.championship[0]} - {self.championship[1]} -> {res}')
        
        print ('')
        print (f'Champion: {res[0]}')
```

```python
tourneys = [kenpom_top_sixteen, net_sixteen, ap_teams, coaches_sixteen, bpi]

teams_agg = dict()

for t in tourneys:
    for idx, team in enumerate(t):
        if team not in teams_agg:
            teams_agg[team] = [idx+1]
        else:
            teams_agg[team].append(idx+1)

print (teams_agg)

teams_avg = dict()
for k, v in teams_agg.items():
    teams_avg[k] = sum(v)/len(v)

{k: v for k, v in sorted(teams_avg.items(), key=lambda item: item[1])}
```
